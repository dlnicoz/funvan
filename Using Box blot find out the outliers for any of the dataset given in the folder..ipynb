```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Placement dataset
placement_df = pd.read_csv('Placement_Data.csv')

# Display basic statistics
print("Basic Statistics:")
print(placement_df.describe())

# Select columns for boxplot
boxplot_columns = ['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p', 'salary']

# Create a boxplot
sns.boxplot(data=placement_df[boxplot_columns])

# Show the plot
plt.title("Boxplot of Numeric Columns")
plt.xlabel("Columns")
plt.ylabel("Values")
plt.show()
```
