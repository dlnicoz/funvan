```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load the Placement dataset
placement_df = pd.read_csv('Placement_Data.csv')

# Display basic statistics
print("Basic Statistics:")
print(placement_df.describe())

# Scatter plot
tips = sns.load_dataset("tips")
sns.scatterplot(x="ssc_p", y="hsc_p", data=placement_df)  # Change x and y labels to ssc_p and hsc_p
plt.title("SSC Percentage vs HSC Percentage")
plt.xlabel("SSC Percentage")
plt.ylabel("HSC Percentage")
plt.show()

# Convert 'workex' column to numeric using label encoding
label_encoder = LabelEncoder()
placement_df['workex_numeric'] = label_encoder.fit_transform(placement_df['workex'])

# Convert 'status' column to numeric using label encoding
placement_df['status_numeric'] = label_encoder.fit_transform(placement_df['status'])

# Drop non-numeric and unnecessary columns
placement_df_numeric = placement_df[['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p', 'status_numeric', 'salary']]

# Compute correlation matrix
corr_matrix = placement_df_numeric.corr()

# Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
